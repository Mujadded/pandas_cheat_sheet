### Multiple column apply

```python
def quality(total_bill, tip):
    if(tip/total_bill > 0.25):
        return 'Generous'
    else:
        return 'Other'
```

```python
quality(16.99,1.01)
```

```python
df['Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'],df['tip']), axis=1)
```

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'], df['tip']) #This is faster
```

```python
df.head()
```

```python
import timeit 
  
# code snippet to be executed only once 
setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv('tips.csv')
def quality(total_bill,tip):
    if tip/total_bill  > 0.25:
        return "Generous"
    else:
        return "Other"
'''
  
# code snippet whose execution time is to be measured 
stmt_one = ''' 
df['Tip Quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
'''

stmt_two = '''
df['Tip Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
'''
  
```

```python
timeit.timeit(setup = setup, 
                    stmt = stmt_one, 
                    number = 1000) 
```

```python
timeit.timeit(setup = setup, 
                    stmt = stmt_two, 
                    number = 1000) 
```

## Merge

##### Inner join

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
# help(pd.merge)
```

```python
pd.merge(registrations, logins, how='inner', on='name')
```

#### left/right merge

```python
pd.merge(left=registrations, right=logins, how='left', on='name')
```

```python
pd.merge(left=registrations, right=logins, how='right', on='name')
```

##### Outer merge

```python
pd.merge(left=registrations, right=logins, how='outer', on='name')
```

```python
registrations = registrations.set_index('name')
```

```python
registrations
```

```python
pd.merge(registrations,logins,left_index=True,right_on='name', how='inner')
```

```python
registrations = registrations.reset_index()
```

```python
registrations
```

```python
registrations.columns = ['reg_name', 'reg_id']
```

```python
registrations
```

```python
results = pd.merge(registrations,logins,left_on='reg_name',right_on='name', how='inner')
```

```python
results.drop('reg_name', axis=1)
```

```python
registrations.columns = ['name','id']
```

```python
logins.columns = ['id','name']
```

```python
pd.merge(registrations,logins,on='name', suffixes=('_reg','_log'))
```
